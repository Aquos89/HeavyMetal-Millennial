## Millennial Generation and Music
The Millennial generation, typically defined as those born between 1981 and 1996, came of age in a period marked by rapid technological change, globalization, and evolving cultural values. As digital natives, Millennials grew up with unprecedented access to media, music platforms, and online communities, which shaped both their modes of cultural consumption and their identity formation. Scholars have noted that Millennials tend to value authenticity, social consciousness, and emotional introspection in artistic expression, often seeking content that resonates with their personal experiences and sociopolitical realities (Howe & Strauss 2000; Bennett 2012). In contrast to previous generations, their relationship with music is less centered on physical ownership or genre loyalty and more on playlists, mood-based listening, and the social context in which music is shared and interpreted.

## Ethnomusicological studies on the thematic content of Heavy Metal
Ethnomusicological approaches to the thematic content of the metal music genre have evolved over time and can be divided into two major schools of thought: a moralistic-critical approach (dominant in the 1980s and 1990s) and a more recent and nuanced sociocultural-academic perspective. The first approach interprets heavy metal as a social threat that encourages dangerous behaviors, especially among youth. It is grounded in theoretical frameworks such as labeling theory, moral panic theory, and the psychology of risk. For this reason, particular emphasis is placed on themes such as sex, drugs, violence, suicide, rebellion, satanism, and antisocial behavior (Arnett 1991, 1993; Ballard & Coates 1995; PMRC 1985; Binder 1993; Bryson 1993, 1996; Took & Weiss 1994).

The second approach views metal as a form of aesthetic, political, and philosophical expression that articulates cultural, personal, and social conflicts—especially those relating to power, death, masculinity, alienation, and the sacred/profane. It is based on the lived experiences of metal composers from each generation, placing particular emphasis on metal as a symbolic language with themes loaded with mythology, politics, religion, and philosophy (Weinstein 2000; Walser 1993; Kahn-Harris 2007; Rafalovich 2006; Moberg 2012; Luhr 2005; Brown 2011; Sylvan 2002).

## On Chabot’s 2012 study
In his master’s thesis titled Thematic Patterns in Millennial Heavy Metal: A Lyrical Analysis, Evan Chabot (2012) set out to systematically analyze the dominant lyrical themes in heavy metal from the 2000s—a period he defines as "Millennial metal." His work is based on the observation that much of the academic literature on heavy metal has been dominated by negative frameworks inherited from the 1980s and 1990s, centered on notions of deviance, moral panic, and psychological harm, while neglecting rigorous analysis of the lyrical content itself, especially in its more recent evolution.

Chabot argues that contemporary heavy metal—like the Millennial generation that produces and consumes it—has undergone significant aesthetic and ideological transformations. Therefore, his central hypothesis is that the lyrics of Millennial metal offer a window into the cultural, social, and personal values of this generation. To test this, he built a sample of 250 English-language songs selected at random from a pool of 200 representative albums released between 2000 and 2009, compiled from specialized rankings and fan forums.

The method consisted of a manual content analysis based on a matrix of 115 lyrical themes, coded according to both literal mention and metaphorical interpretation. The most frequent results were themes of death (31.6%), storytelling/narrative (28.4%), violence (21.2%), social commentary (18.4%), and personal relationships (16%). Through qualitative analysis of how these themes were used in context, Chabot argues that Millennial metal is characterized by an introspective, narrative, and socially aware approach in which metaphor, critical engagement with social structures, and the affirmation of individuality prevail.

## On the topic modelling experiment
To complement and contrast Chabot’s manual classification of lyrical themes in Millennial heavy metal, this project implements a topic modelling experiment using unsupervised machine learning techniques. The aim is to identify thematic patterns in a large corpus of metal lyrics through statistical inference, without relying on predefined categories or interpretive bias. This computational approach enables the discovery of latent structures in the data that may align with, diverge from, or enrich the typologies developed through manual qualitative coding.

For this purpose, the “Large Metal Lyrics Archive” compiled by Alex Mourne and hosted on Kaggle.com was used as the foundational dataset. It contains approximately 220,000 metal song lyrics in multiple languages, along with metadata such as title, artist, album, and year of release. To ensure comparability with Chabot’s 2012 study, the dataset was filtered to include only English-language songs released between 2000 and 2009, resulting in a working corpus of 68,335 lyrics.

The topic modelling method employed was Latent Dirichlet Allocation (LDA), a widely used probabilistic model for uncovering latent topics in large text corpora. LDA assumes that each document (in this case, each song lyric) is a mixture of topics, and each topic is a distribution over words. This model was chosen for its conceptual alignment with the idea of thematic plurality in metal lyrics and for its interpretability in textual data analysis.

To support the interpretation of the resulting topics, a Word2Vec model was also trained on the same corpus. While not a topic model per se, Word2Vec served as a semantic tool to explore the contextual relationships between words, helping to enrich the understanding of each topic's conceptual coherence and sociocultural significance. The combined use of LDA and Word2Vec thus allowed for both structural extraction and semantic interpretation of thematic patterns in Millennial metal lyrics.


## Tokenization and Stopword Removal


```python
#Download the list of English stopwords from NLTK
nltk.download('stopwords')
stop_words = set(stopwords.words('english')) # Load English stopwords into a set for faster lookup


# Define a preprocessing function for lyrics
# - simple_preprocess() tokenizes the text, lowercases it, and removes punctuation
# - deacc=True removes accent marks
# - Then we filter out stopwords
def preprocess_lyrics(lyrics):
    return [word for word in simple_preprocess(lyrics, deacc=True) if word not in stop_words]

# Apply the preprocessing function to all non-empty lyrics
# - dropna() ensures we skip rows with missing lyrics
# - apply() runs the preprocessing on each lyric individually
processed_lyrics = df_en['Lyric'].dropna().apply(preprocess_lyrics)
```

# 4. **Training the Word2Vec Model**



```python
# Convert the processed lyrics (a Series of token lists) into a list of sentences
# Each sentence is a list of tokens (words), already cleaned and preprocessed
sentences = processed_lyrics.tolist()

# Train a Word2Vec model on the list of tokenized lyrics
# Parameters:
# - vector_size=100: dimensionality of the word vectors (embedding size)
# - window=5: maximum distance between the current word and a word within a given context
# - min_count=5: ignore all words with total frequency lower than this (5 times)
# - workers=4: use 4 CPU cores (threads) for training / it reduces time!!
model = Word2Vec(sentences, vector_size=100, window=5, min_count=5, workers=4)
```

```python
# Define a function to explore semantically similar words using the trained Word2Vec model
def explore_word(word, topn=20):
    try:
        # 🔹 model.wv.most_similar(): retrieves words that are semantically close to the input word
        # Parameters:
        #   - word: the target word to explore
        #   - topn: how many of the most similar words to return (default is 20)
        similar_words = model.wv.most_similar(word, topn=topn)

        # 🔹 Print a readable output
        print(f"Words related to '{word}':")
        for sim_word, score in similar_words:
            # Each result is a tuple (word, similarity_score) where the score is between -1 and 1
            print(f"{sim_word} ({score:.2f})")  # Show score rounded to 2 decimal places

    except KeyError:
        # 🔹 If the word was never seen or was too rare during training, Word2Vec will raise a KeyError
        print(f"The word '{word}' is not in the model's vocabulary.")

# Example:
explore_word("corruption")
```

### **Visualization with PCA (Similar Words)**

```python
def visualize_similar_words(word):
    try:
        # Retrieve the top 10 most similar words to the input word
        # model.wv.most_similar() finds words that are closest in vector space to the given word
        # Parameters:
        #   - word: the target word to explore
        #   - topn=10: number of most similar words to return
        similar_words = model.wv.most_similar(word, topn=10)

        # Create a list with the input word followed by its similar words
        # We want to visualize the target word along with its neighbors
        words = [word] + [w for w, _ in similar_words]

        # Get the vector (embedding) of each word in the list
        # Each word vector has 100 dimensions (as defined during Word2Vec training with vector_size=100)
        word_vectors = [model.wv[w] for w in words]

        #  Reduce dimensionality from 100D to 2D using PCA
        # PCA (Principal Component Analysis) helps us project high-dimensional data onto a 2D plane for visualization
        # Parameters:
        #   - n_components=2: we want 2 principal components (x and y axes for the scatter plot)
        pca = PCA(n_components=2)
        result = pca.fit_transform(word_vectors)  # result will be a 2D array with shape (len(words), 2)

        # Create a 2D scatter plot using matplotlib
        plt.figure(figsize=(10, 6))  # set the size of the plot (width=10, height=6 inches)
        plt.scatter(result[:, 0], result[:, 1])  # plot all word vectors as points

        # Label each point with the corresponding word
        for i, w in enumerate(words):
            plt.annotate(w, xy=(result[i, 0], result[i, 1]))

        # Add a title and display the plot
        plt.title(f"PCA Visualization for '{word}' and its neighbors")
        plt.show()

    except KeyError:
        # 🔹 If the word is not found in the Word2Vec vocabulary, show a message
        print(f"The word '{word}' is not in the model's vocabulary.")

# 📌 Example call:
visualize_similar_words("corruption")
```

# 4.2) **Topic Modelling with LDA**

## a) **With 115 topics**

The number 115 corresponds to the topics identified by Chabot (2012) in academic articles and books for the analysis conducted in his study. Since the goal of this project is to compare two models —one manual and one computational— it was necessary to use the same 115 topics in order to assess how well each topic is represented in both approaches.

```python
from nltk.corpus import stopwords
from gensim.utils import simple_preprocess

stop_words = set(stopwords.words('english'))

def preprocess(text):
    return [token for token in simple_preprocess(text, deacc=True) if token not in stop_words]

# Preprocess all lyrics
processed_lyrics = df_en['Lyric'].dropna().apply(preprocess)
```

```python
from gensim.corpora import Dictionary
```

```python
# Create dictionary
dictionary = Dictionary(processed_lyrics)

# Filter out very rare words
dictionary.filter_extremes(no_below=10, no_above=0.5)

# Create corpus
corpus = [dictionary.doc2bow(text) for text in processed_lyrics]
```

```python
from gensim.models.ldamodel import LdaModel
```

```python
# Number of topics to extract from the corpus
# You can adjust this based on your needs (e.g., 50, 100, 115, etc.)
num_topics = 115

# Train the LDA model using Gensim's LdaModel
lda_model115 = LdaModel(
    corpus=corpus,              # The document-term matrix (Bag-of-Words format)
    id2word=dictionary,         # Mapping of word IDs to actual words (i.e., the vocabulary)
    num_topics=num_topics,      # Number of latent topics to be discovered
    random_state=42,            # Fixed random seed for reproducibility
    passes=10,                  # Number of passes through the entire corpus during training
    alpha='auto',               # Let the model learn the document-topic density automatically
    per_word_topics=True        # Enables the model to report per-word topic assignments (useful for analysis)
)
```

```python
# Loop through all topics in the LDA model (from 0 to num_topics - 1)
for idx in range(lda_model115.num_topics):

    # Print the topic header with the topic number
    print(f"🧵 Topic {idx}:")

    # Get the top 50 most representative words for the topic
    # - Each term comes with a probability indicating its relevance to the topic
    terms = lda_model115.show_topic(idx, topn=50)  # topn=50 means we want 50 top words

    # Loop through each word and its associated probability
    for word, prob in terms:
        # Print each word followed by its probability, rounded to 3 decimals
        # All words are printed on the same line with double spaces
        print(f"{word} ({prob:.3f})", end='  ')

    # Print a separator line after each topic for better readability
    print("\n" + "-"*100 + "\n")
```

### Assigning a title to the 115 topics

A function is created that takes the most important words from an LDA topic and searches for a title that summarizes its meaning using a Word2Vec model.

```python
# Define a function to generate a semantic title for an LDA topic
def generate_semantic_topic_title(topic_words, w2v_model, topn=3):
    vectors = []  # Create a list to store the Word2Vec vectors of the topic words

    # Loop through the most representative words of the topic
    for word, _ in topic_words:
        # Check if the word exists in the Word2Vec model's vocabulary
        if word in w2v_model.wv:
            # If yes, retrieve its vector and append it to the list
            vectors.append(w2v_model.wv[word])

    # If no vectors were collected (none of the words were in the model), return a default label
    if not vectors:
        return "Unlabeled"

    # Calculate the average vector (semantic centroid) of all the word vectors in the topic
    avg_vector = sum(vectors) / len(vectors)

    # Use the Word2Vec model to find the top-N most similar words to this centroid vector
    similar_words = w2v_model.wv.similar_by_vector(avg_vector, topn=topn)

    # Combine the most similar words into a string to serve as the semantic title
    title = " / ".join([w for w, _ in similar_words])

    return title  # Return the generated title
```

```python
# Dictionary to store the generated semantic titles
semantic_titles = {}

# Iterate on all topics of the LDA model
for topic_id in range(lda_model115.num_topics):
    # Obtain the 20 most representative words of the topic.
    topic_words = lda_model115.show_topic(topic_id, topn=20)

    # Generating a semantic title using Word2Vec
    title = generate_semantic_topic_title(topic_words, model)

    # Save the title in the dictionary
    semantic_titles[topic_id] = title
```

### Assigning a title to the 10 topics

A function is created that takes the most important words from an LDA topic and searches for a title that summarizes its meaning using a Word2Vec model.

```python
# Define a function to generate a semantic title for an LDA topic
def generate_semantic_topic_title10(topic_words, w2v_model, topn=3):
    vectors = []  # Create a list to store the Word2Vec vectors of the topic words

    # Loop through the most representative words of the topic
    for word, _ in topic_words:
        # Check if the word exists in the Word2Vec model's vocabulary
        if word in w2v_model.wv:
            # If yes, retrieve its vector and append it to the list
            vectors.append(w2v_model.wv[word])

    # If no vectors were collected (none of the words were in the model), return a default label
    if not vectors:
        return "Unlabeled"

    # Calculate the average vector (semantic centroid) of all the word vectors in the topic
    avg_vector10 = sum(vectors) / len(vectors)

    # Use the Word2Vec model to find the top-N most similar words to this centroid vector
    similar_words10 = w2v_model.wv.similar_by_vector(avg_vector10, topn=topn)

    # Combine the most similar words into a string to serve as the semantic title
    title10 = " / ".join([w for w, _ in similar_words10])

    return title10  # Return the generated title
```

```python
# Calculate the average vectors for each semantic category
category_vectors = {}  # Dictionary to store the vectors of each category

# Iterate over each category and its list of keywords
for category, keywords in category_keywords.items():

    # Get the Word2Vec vectors only for words that are in the model's vocabulary
    vectors = [model.wv[word] for word in keywords if word in model.wv]

    # If there is at least one valid vector, calculate the mean to represent the category
    if vectors:
        category_vectors[category] = np.mean(vectors, axis=0)
    else:
        # If no keywords are in the model, assign a zero vector
        category_vectors[category] = np.zeros(model.vector_size)
```

## Define function to vectorize songs

```python
import numpy as np  # Import NumPy for numerical computations (averages, vectors, etc.)

# Define a function that converts a song into an average semantic vector
def document_vector(tokens, model):
    """
    Calculates the average vector of a song based on its words (tokens)
    Parameters:
    - tokens: list of words (already preprocessed) from a song
    - model: Word2Vec model trained on the corpus

    Returns:
    - NumPy vector of size 'vector_size' (dimension of the Word2Vec space)
    """

    # Extract the Word2Vec vector for each word, if it's in the model's vocabulary
    vectors = [model.wv[word] for word in tokens if word in model.wv]

    # If the song has at least one word recognized by the model, average their vectors

    if vectors:
        return np.mean(vectors, axis=0)  # average across words
    else:
        # If no word is in the vocabulary, return a zero vector
        return np.zeros(model.vector_size)
```

### Assignment: Clasify new song

```python
from gensim.utils import simple_preprocess
import numpy as np
from numpy import dot
from numpy.linalg import norm

# Helper function to compute cosine similarity between two vectors
def cosine_similarity(a, b):
    return dot(a, b) / (norm(a) * norm(b) + 1e-10)

# Main function: classifies a lyrics string by theme using Word2Vec and category vectors
def classify_lyrics_by_theme(lyrics, model, category_vectors):
    # Preprocess the lyrics into clean tokens
    tokens = simple_preprocess(lyrics)
    # Retrieve Word2Vec vectors only for words in the model's vocabulary
    vectors = [model.wv[word] for word in tokens if word in model.wv]

    # Compute the average vector of the lyric
    if vectors:
        lyric_vector = np.mean(vectors, axis=0)
    else:
        lyric_vector = np.zeros(model.vector_size)

    # Compute cosine similarity between the lyric vector and each category vector
    similarities = {
        category: cosine_similarity(lyric_vector, cat_vec)
        for category, cat_vec in category_vectors.items()
    }

    # Sort categories by similarity score (highest to lowest)
    sorted_similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)

    # Print the results in a clear format
    for cat, score in sorted_similarities:
        print(f"{cat} → {round(score, 2)}")

    #Return sorted list of (category, similarity) pairs
    return sorted_similarities
```

This project presents a significant methodological comparison between two approaches to the thematic analysis of heavy metal lyrics produced by the Millennial generation. On the one hand, it considers the manual qualitative study conducted by Evan Chabot in 2012, based on an interpretive coding of 250 songs. On the other, it implements an automated analysis using topic modelling techniques (LDA) and semantic similarity models (Word2Vec), applied to a corpus of more than 68,000 English-language songs released between 2000 and 2009.

The rigorous delimitation of the corpus allows for a valid comparison between both methods, strengthening the relevance of the contrast. Both approaches agree that Millennial metal lyrics tend to be introspective, narrative, and socially critical. Despite this, themes such as war and violence remain central and are often presented through metaphors, symbolism, and philosophical allusions.
